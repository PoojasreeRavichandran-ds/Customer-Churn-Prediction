```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(x='Churn_Yes',data=df)
plt.title("Customer Churn Prediction")
plt.show()
```

```python
sns.countplot(x='gender_Male',hue='Churn_Yes',data=df)
plt.title("Gender VS Churn")
plt.show()
```

```python
sns.countplot(x='Partner_Yes',hue='Churn_Yes',data=df)
plt.title("Partner VS Churn")
plt.show()
```

```python
sns.countplot(x='Dependents_Yes',hue='Churn_Yes',data=df)
plt.title("Dependents VS Churn")
plt.show()
```

```python
df['tenure']=pd.cut(df['tenure'],bins=[0, 12, 24, 48, 72],labels=['0-1 yr', '1-2 yr', '2-4 yr', '4-6 yr'])
```

```python
sns.countplot(x='tenure',hue='Churn_Yes',data=df)
plt.title("Tenure VS Churn")
plt.show()
```

```python
sns.countplot(x='Contract_One year',hue='Churn_Yes',data=df)
plt.title("Contract(One Year) VS Churn")
plt.show()
```

```python
sns.countplot(x='Contract_Two year',hue='Churn_Yes',data=df)
plt.title("Contract(Two Year) VS Churn")
plt.show()
```

```python
df['PaymentMethod'] = df[['PaymentMethod_Credit card (automatic)',
                           'PaymentMethod_Electronic check',
                           'PaymentMethod_Mailed check']].idxmax(axis=1)
```

```python
sns.countplot(x='PaymentMethod',hue='Churn_Yes',data=df)
plt.title("Payment Method VS Churn")
plt.xticks(rotation=20)
plt.show()
```
